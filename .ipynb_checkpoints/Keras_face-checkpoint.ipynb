## Face Detection-MTCNN

```python
from matplotlib import pyplot as plt
from matplotlib.patches import Rectangle
from matplotlib.patches import Circle
from PIL import Image 
from numpy import savez_compressed
from numpy import asarray
from os import listdir
from mtcnn.mtcnn import MTCNN
```

```python
%matplotlib inline
```

```python
def extract_image(image):
    img1 = Image.open(image)
    img1 = img1.convert('RGB')
    pixels = asarray(img1)
    detector = MTCNN()
    
    f = detector.detect_faces(pixels)
    #fecthing cordinates
    
        
    x1,y1,w,h = f[0]['box'] #arr[0]
    x1,y1 = abs(x1),abs(y1)
    x2 = abs(x1+w)
    y2 = abs(y1+h)
#locate coordinates of face in image

    store_face = pixels[y1:y2,x1:x2]
    plt.imshow(store_face)
    print("store",store_face)
    image_1 = Image.fromarray(store_face,'RGB')
    image_1  = image_1.resize((160,160))
    face_array = (asarray(image_1))
    return face_array
    
```

```python
#method to fetch the face
def load_faces(directory):
    face =[]
    i=1
    for filename in listdir(directory):
        path = directory+filename
        print(path)
        faces = extract_image(path)
        face.append(faces)
    return face
```

```python
#method to get array of face data and its labels

def load_dataset(directory):
    x,y = [],[]
    i =1
    for subdir in listdir(directory):
        path = directory + subdir + '/'
        #load all faces in subdir
        faces = load_faces(path)
        #create labels
        labels = [subdir for _ in range(len(faces))]
        print("%d there are %d images in the class %s"%(i,len(faces),subdir))
        x.extend(faces)
        y.extend(labels)
        print(y)
        i = i+1
        print(faces)
    return asarray(x),asarray(y)
```

```python
#load the datasets
trainX,trainY = load_dataset('dataset/')
print(trainX.shape,trainY.shape)
#compress the data
savez_compressed('dataset.npz',trainX,trainY)
```

## Featur Extraction-Facenet-Keras

```python
from numpy import load
from numpy import asarray
from numpy import expand_dims
from numpy import savez_compressed
from numpy import reshape
from keras.models import load_model 
```

```python
import keras
```

```python
## generalizing the data and extracting the embeddings

def extract_embeddings(model,face_pixels):
    face_pixels = face_pixels.astype('float32')
    mean = face_pixels.mean()
    std = face_pixels.std()
    face_pixels = (face_pixels-mean)/std
    samples = expand_dims(face_pixels,axis=0)
    yhat = model.predict(samples)
    return yhat[0]
```

```python
import tensorflow as tf
```

```python
#load the compressed data set and facenet model
data = load('dataset.npz')
trainx ,trainy = data['arr_0'],data["arr_1"]
print(trainx.shape,trainy.shape)
model = tf.keras.models.load_model('facenet_keras.h5',compile=False)


#getting the face embeddings
new_trainx = []
for train_pixels in trainx:
    embeddings = extract_embeddings(model,train_pixels)
    new_trainx.append(embeddings)
new_trainx = asarray(new_trainx)
print(new_trainx.shape)

#compress the 128 embeddings of each face
savez_compressed('dataset.npz',new_trainx,trainy)
```

```python
# draw an image with detected objects
def draw_facebox(filename, result_list):
    # load the image
    data = plt.imread(filename)
    # plot the image
    plt.imshow(data)
    # get the context for drawing boxes
    ax = plt.gca()
    # plot each box
    for result in result_list:
    # get coordinates
    x, y, width, height = result['box']
    # create the shape
    rect = plt.Rectangle((x, y), width, height,fill=False, color='orange')
    # draw the box
    ax.add_patch(rect)
    # draw the dots
    for key, value in result['keypoints'].items():
    # create and draw dot
    dot = plt.Circle(value, radius=20, color='red')
    ax.add_patch(dot)
    # show the plot
    plt.show()
```
